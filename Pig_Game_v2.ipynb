```python
import random
import sys
'''Set dice'''
def roll():
    min_value=1
    max_value=6
    roll=random.randint(min_value,max_value)
    return roll



'''set number of players input and check the input'''
while True:
    players=input("No of players(2-4)? ")
    if players.isdigit():
        players=int(players)
        if 2<= players <=4:
            break
        else:
            print("Must be between 2-4")
    else:
        print("Invalid number,try again")



'''set max score'''
max_score=50



'''set strating score as 0 for all players'''
player_score= [0 for i in range(players)]



'''set a check for max player score less than and greater than'''
while max(player_score)<max_score:
    for i in range(players):
        print("\nPlayer number",i+1,"turn has just started!")
        print("Your total score is: ",player_score[i],"\n")
        current_score=0

        while True:
            rolling=input("Do you want to roll the dice(y for yes/ n for no/ e for exit)?: ")
            if rolling.lower()== "e":
                max_score= max(player_score)
                winning_player=player_score.index(max_score)
                print ("player_number", winning_player+1, " is the winner with a score of: ", max_score)
                break
                exit()
                
            elif rolling.lower()== "y":
                value=roll()
                if value==1:
                    print("You rolled a 1! Turn Over!")
                    current_score=0
                    player_score[i]*=0
                    break
                    
                elif max(player_score)+value>50:
                    print("You rolled a: ",value)
                    print("You have rolled over a total of 50! Start from 0")
                    current_score=0
                    player_score[i]*=0
                    break

                elif max(player_score)== max_score:
                    print("You rolled a: ", value)
                    print("Your total score is: ", player_score[i])
                    print("your total score has reached max score. If I were you I would stick to no(n) to roll")
                    break
                
                else:
                    current_score+=value
                    player_score[i]+=value
                    print("you rolled a: ",value)
                    print("your score in this round is: ",current_score)
                    print("your total score is: ",player_score[i])
                    
            elif rolling.lower()== "n":
                current_score+=0
                print("your total score is: ",player_score[i])
                break
                
            else:
                print("Incorrect option selected")
```
